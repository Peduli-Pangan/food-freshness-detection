```python
from icrawler.builtin import GoogleImageCrawler

google_crawler = GoogleImageCrawler(
    feeder_threads=1,
    parser_threads=2,
    downloader_threads=4,
    storage={'root_dir': 'train_set'}
)

filters = dict(
    date=((2020, 1, 1), (2024, 12, 30))
)
google_crawler.crawl(keyword='', filters=filters, max_num=500, file_idx_offset=0)
```
